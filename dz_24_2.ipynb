### Завдання 2.

Є можливість вибрати спосіб виробництва і реалізації двох наборів товарів широкого вжитку. За даними відділу маркетингу, яким були проведені дослідження ринку, можливий прибуток від виробництва і реалізації Х і Y наведено в таблицях (Х, Y – прибуток у грошових одиницях). Потрібно оцінити ступінь ризику і запропонувати рішення щодо випуску і реалізації одного із наборів товарів.

| X      | 1000    | 1500    | 2000    |                  
| :---   | :---:   | :---:   | ---:    |                  
| P      | 0.5     | 0.3     | 0.2     |                  

| Y      | 1000    | 1500    | 1750    |
| :---   | :---:   | :---:   | ---:    | 
| P      | 0.4     | 0.4     | 0.2     |

### Вирішення

```python
x1 = 1000
x2 = 1500
x3 = 2000
P_x1 = 0.5
P_x2 = 0.3
P_x3 = 0.2
y1 = 1000
y2 = 1500
y3 = 1750
P_y1 = 0.4
P_y2 = 0.4
P_y3 = 0.2

# Знаходимо математичне очікування для кожного варіанту M(X) та M(Y)
M_X = x1 * P_x1 + x2 * P_x2 + x3 * P_x3
M_Y = y1 * P_y1 + y2 * P_y2 + y3 * P_y3
print(f"Математичне очікування для кожного варіанту - {M_X}, {M_Y}")

# Знаходимо дисперсію кожного варіанту -D(X) та D(Y)
D_X = x1 ** 2 * P_x1 + x2 ** 2 * P_x2 + x3 ** 2 * P_x3 - M_X ** 2
D_Y = y1 ** 2 * P_y1 + y2 ** 2 * P_y2 + y3 ** 2 * P_y3 - M_Y ** 2
print(f"Дисперсія для кожного варіанту - {D_X}, {D_Y}")

# Знаходимо середньоквадратичне відхилення для SD для X та Y
SD_X = D_X ** 0.5
SD_Y = D_Y ** 0.5
print(f"Середньоквадратичне відхилення для кожного варіанту - {SD_X}, {SD_Y}")

if SD_X > SD_Y:
    print("Пропонується до реалізації другий варіант, у якого ризик менший")
else:
    print("Пропонується до реалізації перший варіант, у якого ризик менший")
```
